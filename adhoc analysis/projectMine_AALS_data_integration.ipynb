```python
import pandas as pd
```

```python
aalsGenotypesPath = "Variant_report_NUPs_fixed_2022-03-28.xlsx"
mineGenotypesPath = "nup_rsids_project_mine.tsv"
```

# Parse genotype dataframes

```python
aalsGenotypes = pd.read_excel(aalsGenotypesPath, sheet_name="all cases vs all controls").dropna(subset="position")
aalsGenotypes['position'] = aalsGenotypes['position'].astype(int)
aalsGenotypes = aalsGenotypes.set_index(['chrom', 'position', 'rsID', 'Gene'])

mineGenotypes = pd.read_csv(mineGenotypesPath, sep="\t").rename(columns={'Unnamed: 0': 'rsID'}).set_index('rsID').T
```

```python
aalsGenotypes
```

```python
mineGenotypes
```

# Determine common variants

```python
aalsGenotypesWithMineRSids = aalsGenotypes.loc[aalsGenotypes.index.droplevel(['chrom', 'position', 'Gene']).isin(mineGenotypes.index)]
```

```python
aalsGenotypesWithMineRSids.index
```

```python
chromPosRSidGeneIndex = aalsGenotypesWithMineRSids.index.tolist()
mineRSidIndex = mineGenotypes.index.tolist()
mineChromPosRSidGeneList = []

for rsid in mineRSidIndex:
    for i, multiIndexItem in enumerate(chromPosRSidGeneIndex):
        if multiIndexItem[2] == rsid:
            mineChromPosRSidGeneList.append(multiIndexItem)
            chromPosRSidGeneIndex.remove(multiIndexItem)
```

```python
mineGenotypes.index = pd.MultiIndex.from_tuples(mineChromPosRSidGeneList, names=['chrom', 'position', 'rsID', 'Gene'])
```

```python
mineGenotypes
```

## Merge common Project MinE and AALS variants

```python
# trim summary stat cols
aalsGenotypes = aalsGenotypes.iloc[:,:-10].loc[mineChromPosRSidGeneList]
mergedGenotypes = pd.concat([aalsGenotypes, mineGenotypes], axis=1)
```

```python
mergedGenotypes
```

```python
mergedGenotypes.to_csv("NUP_aals_mine_genotypes.tsv", sep="\t")
```

# Save MinE case/control IDs

```python
mineCaseIDs, mineControlIDs = [], []

for sampleId in mineGenotypes.columns:
    if '1_' == sampleId[:2]:
        mineCaseIDs.append(sampleId)
    elif '0_' == sampleId[:2]:
        mineControlIDs.append(sampleId)
```

```python
pd.Series(mineCaseIDs, name='id').to_csv("mine_case_ids.tsv", index=False)
pd.Series(mineControlIDs, name='id').to_csv("mine_control_ids.tsv", index=False)
```
